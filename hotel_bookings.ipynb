# IMPORTING THE LIBRARIES AND LOADING THE DATASET

```python
# IMPORTING THE LIBRARIES

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# LOADING THE DATASET

df=pd.read_csv("hotel_bookings.csv")
df.head()
```

# EXPLORATARY DATA ANALYSIS

```python
# CHECKING THE NUMBER OF ROWS AND COLUMNS
df.shape
```

```python
# DIFFERENT COLUMNS PRESENT 

df.columns
```

```python
# CHECKING THE DATATYPE OF COLUMNS

df.info()
```

```python
# CHANGING THE DATATYPE OF 'reservation_status_date' COLUMN FROM 'object' TO 'datetime64[ns]'

df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
# USING THE DESCRIBE FUNCTION FOR THE 'object' TYPE COLUMNS

df.describe(include='object')
```

```python
# SEEING ALL THE COLUMNS WITH ALL THEIR UNIQUE VALUES PRESENT 

for cols in df.describe(include='object').columns:
    print(cols)
    print(df[cols].unique())
    print('-'*100)
```

```python
# CHECKING ALL THE NULL VALUES PRESENT IN EACH COLUMNS

df.isnull().sum()
```

```python
# DROPPING THE COLUMNS AND ALSO SOME OF THE NULL DATA

df.drop(['agent','company'],axis=1,inplace=True)
df.dropna(inplace=True)
```

```python
# USING THE DESCRIBE FUNCTION FOR ALL NUMERICAL COLUMNS

df.describe()
```

```python
# KEEPING ONLY THOSE ROWS WHOSE 'adr' COLUMN VALUE IS LESS THAN 5000

df=df[df['adr']<5000]
```

# DATA VISUALIZATIONS

```python
# CHECKING THE PERCENTAGE OF BOOKINGS CANCELED AND NOT CANCELED

cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
# CREATING A BAR GRAPH FOR SEEING BOOKINGS CANCELED VS NOT CANCELED

plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts())
plt.title("RESERVATION STATUS COUNT")
plt.show()
```

```python
# CREATING THE COUNTPLOT FOR THE RESERVATION STATUS IN DIFFERENT HOTELS

ax1=sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title("RESERVATION STATUS IN DIFFERENT HOTELS")
plt.xlabel("Hotel")
plt.ylabel("Number of Reservations")
plt.legend(["not canceled","canceled"])
plt.show()
```

```python
# CHECKING THE PERCENTAGE OF BOOKINGS CANCELED AND NOT CANCELED OF RESORT HOTEL

resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# CHECKING THE PERCENTAGE OF BOOKINGS CANCELED AND NOT CANCELED OF CITY HOTEL

city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# GROUPING THE DATA ON THE BASIS OF 'reservation_status_date' COLUMN

resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# CREATING A PLOT ON THE BASIS OF CITY HOTEL AND RESORT HOTEL FOR THE AVERAGE DAILY RATE

plt.figure(figsize=(18,8))
plt.plot(resort_hotel.index,resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.title("AVERAGE DAILY RATE IN CITY AND RESORT HOTEL")
plt.legend()
plt.show()
```

```python
# CREATING A COUNTPLOT FOR THE 'month' COLUMN ON THE BASIS OF 'is_canceled' COLUMN

plt.figure(figsize=(14,6))
df['month']=df['reservation_status_date'].dt.month
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.title("RESERVATION STATUS PER MONTH")
plt.xlabel("Month")
plt.ylabel("Number of Reservations")
plt.legend(["Not Canceled","Canceled"])
plt.show()
```

```python
# CREATING A BAR PLOT FOR THE 'month' COLUMN AND THE 'adr' COLUMN

plt.figure(figsize=(14,6))
ax1=sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title("ADR PER MONTH")
plt.show()
```

```python
# CREATING A PIECHART FOR THE TOP 10 COUNTRIES WITH RESERVATION CANCELED

plt.figure(figsize=(6,5))
canceled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]
plt.title("TOP 10 COUNTRIES WITH RESERVATION CANCELED")
plt.pie(top_10_country,autopct='%.2f',labels=top_10_country.index)
plt.show()
```

```python
# PERCENTAGE OF BOOKING FROM DIFFERENT FACILITY

df['market_segment'].value_counts(normalize=True)
```

```python
# PERCENTAGE OF CANCELATION FROM DIFFERNET FACILITY

canceled_data['market_segment'].value_counts(normalize=True)
```

```python
# CREATING A PLOT ON AVERAGE DAILY RATE AND SEE THE THE NUMBER OF CANCELETION AND NON-CANCELETION FOR ALL DATES

canceled_data_adr=canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_data_adr.reset_index(inplace=True)
canceled_data_adr.sort_values('reservation_status_date',inplace=True)

not_canceled_data=df[df['is_canceled']==0]
not_canceled_data_adr=not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data_adr.reset_index(inplace=True)
not_canceled_data_adr.sort_values('reservation_status_date',inplace=True)

plt.figure(figsize=(20,10))
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label="not canceled")
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label="canceled")
plt.title("AVERAGE DAILY RATE")
plt.legend()
plt.show()
```

```python
# TAKING THE DATA ONLY FROM THE YEAR 2016 TO 9TH MONTH OF 2017

canceled_data_adr=canceled_data_adr[(canceled_data_adr['reservation_status_date']>'2016') & (canceled_data_adr['reservation_status_date']<'2017-09')]
not_canceled_data_adr=not_canceled_data_adr[(not_canceled_data_adr['reservation_status_date']>'2016') & (not_canceled_data_adr['reservation_status_date']<'2017-09')]
```

```python
# CREATING A PLOT ON AVERAGE DAILY RATE AND SEE THE THE NUMBER OF CANCELETION AND NON-CANCELETION FOR THE FILTERED DATE RANGE

plt.figure(figsize=(20,10))
plt.plot(not_canceled_data_adr['reservation_status_date'],not_canceled_data_adr['adr'],label="not canceled")
plt.plot(canceled_data_adr['reservation_status_date'],canceled_data_adr['adr'],label="canceled")
plt.title("AVERAGE DAILY RATE")
plt.legend()
plt.show()
```
